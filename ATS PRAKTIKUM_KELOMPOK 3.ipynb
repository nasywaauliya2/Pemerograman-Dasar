```python
#2. Fungsi / Functions

# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Bonus: Contoh cara memanggil fungsi
input_energi = 500
output_energi = 400

hasil = hitung_efisiensi(input_energi, output_energi)
print("Efisiensi Turbin =", hasil)
```

Program ini dibuat untuk menghitung efisiensi turbin dengan memanfaatkan fungsi agar perhitungan lebih rapi dan mudah digunakan kembali. Fungsi hitung_efisiensi menerima dua nilai, yaitu energi masuk dan energi keluar, kemudian menghitung efisiensi dengan rumus output/input. Setelah itu program memberikan contoh nilai energi masuk 500 dan energi keluar 400, sehingga hasil efisiensinya adalah 0.8. Program ini membantu menyelesaikan masalah perhitungan efisiensi agar tidak perlu dilakukan secara manual dan mengurangi kemungkinan kesalahan hitung.

```python
#6. Variabel & Ekspresi

# Program menghitung daya listrik (P = V x I)

# Input nilai tegangan dan arus
tegangan = 220
arus = 15

# Menghitung daya listrik
daya = tegangan * arus

# Menampilkan hasil perhitungan
print("Tegangan (V) =", tegangan, "Volt")
print("Arus (I) =", arus, "Ampere")
print("Daya listrik (P) =", daya, "Watt")
```

Program ini dibuat untuk menghitung daya listrik menggunakan rumus P = V × I, yang mana P adalah daya dalam Watt, V adalah tegangan dalam Volt, sedangkan I adalah arus dalam Ampere. Pada program ini mendefinisikan dua variabel yaitu tegangan = 220 dan untuk arus yaitu = 15, kemudian proses perhitungan daya dengan mengalikan kedua nilai tersebut, dan hasilnya disimpan ke dalam variabel daya. Setelah perhitungan selesai, program menampilkan nilai tegangan, arus, dan daya listrik ke layar menggunakan perintah print(). Maka outputnya yaitu 3300 watt. Program ini membantu meneyelesaikan masalah erhitungan daya listrik secara otomatis sehingga tidak perlu menghitung manual, dan dapat mempermudah untuk memahami konep pengunaan variabel dan operasi aritmatika dalam pemrograman.

```python
#8. Fungsi / Modularisasi

# Program menghitung efisiensi termal turbin gas
# Rumus: Efisiensi (%) = (energi keluar / energi masuk) × 100

def hitung_efisiensi_termal():
    # Meminta input dari pengguna
    energi_masuk = float(input("Masukkan energi masuk (dalam Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (dalam Joule): "))

    # Menghitung efisiensi
    efisiensi = (energi_keluar / energi_masuk) * 100

    # Menampilkan hasil
    print("\n=== HASIL PERHITUNGAN ===")
    print(f"Energi masuk   : {energi_masuk} Joule")
    print(f"Energi keluar  : {energi_keluar} Joule")
    print(f"Efisiensi termal: {efisiensi:.2f}%")

# Memanggil fungsi
hitung_efisiensi_termal()
```

program ini untuk menghitung efisiensi termal pada turbin gas dengan menggunakan konsep fungsi agar lebih terstruktur dan mudah dipahami. Dalam program ini dibuat fungsi bernama hitung_efisiensi_termal() untuk menginput berupa energi masuk dan keluar satuan Joule, Nilai input tersebut diubah menjadi tipe data float agar dapat digunakan dalam perhitungan. Setelah perhitungan dilakukan, program menampilkan hasil berupa energi masuk, energi keluar, serta nilai efisiensi termal dalam bentuk persentase dengan dua angka di belakang koma. Dengan nilai masuk 8000 J dan energi keluar 4200 J maka outputnya keluar dengan persentase 52,2%. Program ini menghasilkan informasi mengenai seberapa efektif energi yang digunakan diubah menjadi output pada sistem turbin gas. Masalah yang diselesaikan oleh program ini adalah bagaimana menghitung efisiensi termal dengan cepat dan akurat tanpa harus melakukan perhitungan manual, sehingga memudahkan pengguna dalam menganalisis kinerja sistem termal.

```python
#10. While Loop & Logika

# Inisialisasi volume awal tangki
tangki = 0

# Loop selama tangki belum penuh
while tangki < 100:
    tangki += 10  # naikkan 10 liter tiap iterasi
    print(f"Volume tangki saat ini: {tangki} liter")

# Logika kondisi peringatan
if tangki == 80:
      print("WARNING: Hampir penuh!")
elif tangki == 100:
      print("TANGKI PENUH – Pengisian berhenti.")
```

Program di atas berfungsi untuk proses pengisian tangki bahan bakar menggunakan perulangan while dan logika if-elif. Pada awalnya, nilai tangki ditetapkan 0 liter. Selama volume tangki masih kurang dari 100 liter, program akan terus menambah 10 liter setiap kali perulangan dijalankan dan menampilkan volume tangki saat itu. Setelah proses pengisian selesai, program akan mengecek kondisi tangki. Jika volume mencapai 80 liter, akan muncul peringatan “WARNING: Hampir penuh!”, dan jika sudah 100 liter, program menampilkan pesan “TANGKI PENUH Pengisian berhenti.” Secara sederhana, program ini menunjukkan bagaimana pengulangan dan pengecekan kondisi digunakan untuk mengontrol jalannya suatu proses

```python
#Challenge

# Fungsi untuk menghitung energi listrik (P = V x I x t)
def hitung_energi(tegangan, arus, waktu):
    energi = tegangan * arus * waktu
    return energi

# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = (output_energi / input_energi) * 100
    return efisiensi

print("=== PROGRAM PERHITUNGAN ENERGI & EFISIENSI TURBIN ===")

jumlah_turbin = int(input("Masukkan jumlah turbin yang ingin dihitung: "))

# Menggunakan perulangan for untuk menghitung beberapa turbin
for i in range(jumlah_turbin):
    print(f"\n--- Turbin ke-{i+1} ---")
    tegangan = float(input("Masukkan tegangan (Volt): "))
    arus = float(input("Masukkan arus (Ampere): "))
    waktu = float(input("Masukkan waktu operasi (detik): "))

    input_energi = hitung_energi(tegangan, arus, waktu)
    output_energi = float(input("Masukkan output energi turbin (Joule): "))

    efisiensi = hitung_efisiensi(input_energi, output_energi)
    print(f"Efisiensi Turbin = {efisiensi:.2f}%")

    # Percabangan untuk menilai efisiensi
    if efisiensi >= 90:
        print("Kinerja: Sangat Baik")
    elif efisiensi >= 75:
        print("Kinerja: Baik")
    elif efisiensi >= 50:
        print("Kinerja: Cukup")
    else:
        print("Kinerja: Buruk")

# Contoh perulangan while untuk menanyakan apakah ingin menghitung lagi
ulang = input("\nApakah ingin menghitung lagi? (ya/tidak): ").lower()
while ulang == "ya":
    print("\n=== Perhitungan Ulang ===")
    input_energi = float(input("Masukkan input energi (Joule): "))
    output_energi = float(input("Masukkan output energi (Joule): "))
    efisiensi = hitung_efisiensi(input_energi, output_energi)
    print(f"Efisiensi = {efisiensi:.2f}%")
    ulang = input("\nApakah ingin menghitung lagi? (ya/tidak): ").lower()

print("\nProgram selesai. Terima kasih telah menggunakan sistem ini!")
```

Program ini digunakan untuk menghitung energi dan efisiensi turbin pada sistem pembangkit listrik. Di dalamnya:

Fungsi digunakan untuk memisahkan bagian perhitungan energi dan efisiensi agar kode lebih rapi.

Percabangan (if, elif, else) digunakan untuk menilai apakah kinerja turbin termasuk sangat baik, baik, cukup, atau buruk.

Perulangan for digunakan untuk menghitung beberapa turbin sekaligus.

Perulangan while digunakan agar pengguna bisa mengulang perhitungan tanpa keluar dari program.

Secara singkat, program ini membantu mengevaluasi kinerja turbin dengan cara interaktif dan efisien.
